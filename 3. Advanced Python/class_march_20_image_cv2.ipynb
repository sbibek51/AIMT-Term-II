```python
import cv2
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('./data/nepal_airline.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image,cmap='gray'),plt.axis('off')
plt.show()
```

```python
image_bgr=cv2.imread('./data/nepal_airline.jpg',cv2.IMREAD_COLOR)
```

```python
image_rgb = cv2.cvtColor(image_bgr,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb),plt.axis('off')
plt.show()
```

```python
plt.imshow(image_bgr),plt.axis('on')
plt.show()
```
