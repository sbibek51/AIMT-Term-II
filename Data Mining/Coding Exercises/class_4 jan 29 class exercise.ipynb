Q1 . write a  python program to compute the sum of the two reversed numbers .
and display the sum in reverse form:
input : 13,`14
output: 27
    note ther reuslt will not be unique for the every numbers for example 31 is a reversed form of 
    several numbers of 13,130,1300 etc. therefore all leading zeros will be ommited.


```python
x= int(input('Enter a first integer number:'))
y= int (input ('Enter a second integer number:'))
print('Given numbers are {} and {}'.format(x,y))

rev1=int(str(x)[::-1])
rev2=int(str(y)[::-1])
rev3= int(str(rev1+rev2)[::-1])

print('reverse of first number is',rev1)
print('reverse of second number is',rev2)
print('sum of two reverse number is:',rev1+rev2)
print('sum of two reverse number in reverse form is:',rev3)
```

Q2.Write a program to check wheather a given number is an ugly number.
input 12
output: True
    ugly numbers are positive numbers whose only prime factors are 2,3, or 5.
    The sequence 1,2,3,4,5,6,8,9,10,12,... shows the first 10 ugly numbers.
    note : 1 is typically treated as an ugly numbers.

```python
def is_ugly(num):
    factors = [2, 3, 5]
    
    if num == 1:
        return True
    
        for factor in factors:
            while num % factor == 0:
                num /= factor
            
    return num == 1

while True:
    x = input('Enter a number to check whether it is an ugly number or not and e to exit: ')
    if x.lower() == 'e':
        break
    else:
        x = int(x)
        result = is_ugly(x)
        if result:
            print('True')
        else:
            print('False')
```
