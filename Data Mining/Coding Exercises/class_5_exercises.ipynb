A Pythagorean triple consists of three positive integers a,b and c , such that a^2 +b^2 =c^2. Such triple is commonly written (a,b,c) and a well known example is (3,4,5). There exists exactly one Pythogorean triplet for which x+y+z =1000.

```python
def pytha_triplet(target_sum):
    for x in range (1,target_sum):
        for y in range (x,target_sum ):
            z= target_sum-x-y
            if (x**2) +y**2==z**2:
                return x,y,z

    return None

# Example usage:
pytha_triplet(1000)
```
