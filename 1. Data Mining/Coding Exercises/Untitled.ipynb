```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import accuracy_score

# Load your dataset
data = pd.read_csv('../data/data march_4 class work.csv')

# Assuming 'LotArea' as predictor and 'SalePrice' as target
X = data[['LotArea']].values
y_true = data['SalePrice'].values

# Perform hierarchical clustering using SciPy
Z = linkage(X, 'ward')

# Plot dendrogram
plt.figure(figsize=(12, 6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram (SciPy)')
plt.xlabel('Data point index')
plt.ylabel('Distance')
plt.show()

# Determine clusters using fcluster from SciPy
max_d = 2000  # Maximum distance threshold to form clusters
clusters = fcluster(Z, max_d, criterion='distance')

# Calculate accuracy using SciPy
accuracy_scipy = accuracy_score(y_true, clusters)
print("Accuracy (SciPy):", accuracy_scipy)

# Perform hierarchical clustering using Scikit-learn
model = AgglomerativeClustering(n_clusters=2, linkage='ward')
clusters_sklearn = model.fit_predict(X)

# Calculate accuracy using Scikit-learn
accuracy_sklearn = accuracy_score(y_true, clusters_sklearn)
print("Accuracy (Scikit-learn):", accuracy_sklearn)
```
