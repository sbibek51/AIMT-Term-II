```python
dataset = [
    ['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yoghurt'],
    ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yoghurt'],
    ['Milk', 'Apple','Kidney Beans', 'Eggs'] ,
    ['Milk', 'Unicorn', 'Corn', 'Kidney Beans','Yoghurt']  ,
    ['Corn', 'Onion', 'Onion', 'Kidney Beans','Ice Cream', 'Eggs']
]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset
```

```python
te =TransactionEncoder()
te_aray = te.fit(dataset).transform(dataset)
```

```python
df =pd.DataFrame(te_aray,columns=te.columns_)
df
```

There are now 11 items in the list and the data encoded is shown in above dataframe

```python
from mlxtend.frequent_patterns import apriori
```

Printing the frequent itemsets with their support

```python
frequent_itemsets = apriori(df,min_support=0.6,use_colnames=False)
frequent_itemsets
```

## Association Rule Mining

```python
from mlxtend.frequent_patterns import association_rules
res = association_rules(frequent_itemsets,metric='confidence', min_threshold=0.8)
```

```python
res
```

```python
res_1 = res[['antecedents','consequents','support','confidence','lift']]
```

```python
res_1
```

```python
res_2 = res_1[res_1['confidence']>=1]
res_2
```

By doing this implementation we have learned how apriori algorithm works clearly.

## Now we are going to implement a FP based algorithm

```python
# looking at the execution time of previous algorithm i.e. Apriori
```

```python
from mlxtend.frequent_patterns import apriori
%timeit apriori (df,min_support=0.6)
```

```python
from mlxtend.frequent_patterns import fpgrowth
%timeit fpgrowth(df,min_support=0.6)
```

## Conclusion

for the same dataset we have applied the apriori and fpgrowth algorithm and after calculating the computing time apriori algorithm took 2.63 ms whereas fpgrowth algorithm takes only 955 microsecond (µs ). Hence we found that fpgrowth algorithm is more efficient in terms of computation time it is way faster than apriori algorithm.
