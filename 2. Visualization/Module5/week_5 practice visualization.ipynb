### ***********************************Bar Graph********************************

1) Vertical bar graph

plt.bar(x,heights,[width])
* x: list of x -co-ords where the bars are centered
*heights : list of heights of each bar
*width: single value , default is 0.8
    
2) Horizontal bar graph : bars are oriented horizontally
plot.barh(......)

```python
plt.bar(['A','B','C','D'],[20,40,10,15],width =0.4)
```

3) ##Grouped Bar Graph
* for a vertical bar graph: Each category in the x-axis has multiple sub categories
* for a horizontal bar graph: Each category in the y-axis has multiple sub categories


```python
np.arange(4)
```

```python
np.arange(1,4)
```

```python
courst_list = ['CSE', 'EXTC', 'MECH','AERO','METAL']
barIndex_alongXaxis=np.arange(len(courst_list))
girlsCount_barHeights =[100,40,60,240,300]
boysCount_barHeights =[200,70,30,120,15]

width_per_bar = 0.4

bar1=barIndex_alongXaxis
bar2=[(i+width_per_bar)  for i in bar1] 

plt.bar(bar1,girlsCount_barHeights,width_per_bar,label ='Girls')
plt.bar(bar2,boysCount_barHeights,width_per_bar,label ='Boys')


plt.xticks((bar1+ (width_per_bar/2)),courst_list)

plt.title('Bar chart of engineering branch genderwise')
plt.xlabel('Number of students')
plt.ylabel('Engineering Branches')
```
